```python
from transformers import AutoTokenizer, AutoModel
import torch
```

```python
#Mean Pooling - Take attention mask into account for correct averaging
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output[0] #First element of model_output contains all token embeddings
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    sum_embeddings = torch.sum(token_embeddings * input_mask_expanded, 1)
    sum_mask = torch.clamp(input_mask_expanded.sum(1), min=1e-9)
    return sum_embeddings / sum_mask
```

```python
#Sentences we want sentence embeddings for
sentences = ['Привет! Как твои дела?',
             'А правда, что 42 твое любимое число?']
```

```python
#Load AutoModel from huggingface model repository
tokenizer = AutoTokenizer.from_pretrained("sberbank-ai/sbert_large_nlu_ru")
model = AutoModel.from_pretrained("sberbank-ai/sbert_large_nlu_ru")
```

```python
#Tokenize sentences
encoded_input = tokenizer(sentences, padding=True, truncation=True, max_length=24, return_tensors='pt')
encoded_input
```

```python
#Compute token embeddings
with torch.no_grad():
    model_output = model(**encoded_input)
```

```python
#Perform pooling. In this case, mean pooling
sentence_embeddings = mean_pooling(model_output, encoded_input['attention_mask'])
```

```python
import keras
```

```python
model_reg = keras.models.Sequential()
model_reg.add(keras.layers.Dense(52, input_shape=(1024,), activation='relu'))
model_reg.add(keras.layers.Dense(1, activation='linear'))
model_reg.compile(loss='mean_squared_error', optimizer='adam', metrics=['mean_absolute_percentage_error'])

print(model_reg.summary())
```

```python
for i in range(0, 2489, 10):
  encoded_input = tokenizer(X_train[i:i+10], truncation=True, max_length=50, padding=True, return_tensors='pt')
  model_output = model(**encoded_input)
  X_train_prep = mean_pooling(model_output, encoded_input['attention_mask'])
  model_reg.fit(X_train_prep.tolist(), y_train[i:i+10])
```

```python
def predict(X):
  encoded_input = tokenizer(X, truncation=True, max_length=50, padding=True, return_tensors='pt')
  model_output = model(**encoded_input)
  X_prep = mean_pooling(model_output, encoded_input['attention_mask'])
  return model_reg.predict(X_prep.tolist())
```
